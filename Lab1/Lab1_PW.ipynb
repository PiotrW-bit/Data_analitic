```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import beta
```

```python
uniform = np.random.uniform(0, 1, 1000)
normal = np.random.normal(0, 1, 1000)
normal2 = np.random.normal(0, 10, 1000)
beta_1 = np.random.beta(1, 1, 1000)
beta_2 = np.random.beta(2, 2, 1000)
beta_3 = np.random.beta(3, 9, 1000)
normal_1 = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 10000)
normal_2 = np.random.multivariate_normal([0, 0], [[2, 1], [1, 2]], 10000)
```

Creating samples for data using different distributions

```python
def normal_d(x, mu, sigma):
    p = 1 / np.sqrt(2 * np.pi * sigma**2)
    return p * np.exp(-0.5 / sigma**2 * (x - mu)**2)
x_1 = np.linspace(-4, 4, 100)
a_1 = normal_d(x_1, 0, 1)
x_2 = np.linspace(-30, 30, 100)
a_2 = normal_d(x_2, 0, 10)
beta_x = np.linspace(0, 1, 100)
a_3 = beta.pdf(beta_x, 1, 1)
a_4 = beta.pdf(beta_x, 2, 2)
a_5 = beta.pdf(beta_x, 3, 9)

fig, ax = plt.subplots(2, 3, figsize=(15, 10))
ax[0,0].hist(uniform, 20, density=True, label='Samples')
ax[0,0].plot([0, 1],[1, 1], label='Analitical')
ax[0,0].set_title('Uniform distribution on interval [0, 1]')
ax[0,0].legend(loc='upper right')

ax[0,1].hist(normal, 20, density=True, label='Samples')
ax[0,1].plot(x_1,a_1, label='Analitical')
ax[0,1].set_title('Normal distribution $\mu$=0, $\sigma$=1')
ax[0,1].legend(loc='upper right')

ax[0,2].hist(normal2, 20, density=True, label='Samples')
ax[0,2].set_title('Normal distribution $\mu$=0, $\sigma$=10')
ax[0,2].plot(x_2,a_2, label='Analitical')
ax[0,2].legend(loc='upper right')

ax[1,0].hist(beta_1, 20, density=True, label='Samples')
ax[1,0].set_title(r'Beta distribution $\alpha$=1, $\beta$=1')
ax[1,0].plot(beta_x,a_3, label='Analitical')
ax[1,0].legend(loc='upper right')

ax[1,1].hist(beta_2, 20, density=True, label='Samples')
ax[1,1].set_title(r'Beta distribution $\alpha$=2, $\beta$=2')
ax[1,1].plot(beta_x,a_4, label='Analitical')
ax[1,1].legend(loc='upper right')

ax[1,2].hist(beta_3, 20, density=True, label='Samples')
ax[1,2].set_title(r'Beta distribution $\alpha$=3, $\beta$=9')
ax[1,2].plot(beta_x,a_5, label='Analitical')
ax[1,2].legend(loc='upper right')
```

Presentation of data samples on histograms witch where normalised (integral of histogram plot is equal to 1) compared with analitical formulas of functions. We can see that analitical functions match with the density presented on normalised hitograms.

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 15))
ax[0,0].hist2d(normal_1[:,0], normal_1[:,1], 20)
ax[0,0].set_title('2D normal distribution $\mu$=0 $\Sigma$=[[1, 0],[0, 1]] matplotlib')
sns.histplot(x=normal_1[:,0], y=normal_1[:,1], ax=ax[0,1])
ax[0,1].set_title('2D normal distribution $\mu$=0 $\Sigma$=[[1, 0],[0, 1]] seaborn')
ax[1,0].hist2d(normal_2[:,0], normal_2[:,1], 20)
ax[1,0].set_title('2D normal distribution $\mu$=0 $\Sigma$=[[2, 1],[1, 2]] matplotlib')
sns.histplot(x=normal_2[:,0], y=normal_2[:,1], ax=ax[1,1])
ax[1,1].set_title('2D normal distribution $\mu$=0 $\Sigma$=[[2, 1],[1, 2]] seaborn')
```

Presentation of plots for 2D distribution in two different python libraries.

```python
a = np.random.normal(2, 0.1, 1000)
b = np.random.normal(0.2, 0.2, 1000)
x = np.random.normal(0, 3, 1000)
d = np.abs(np.random.normal(0, 1, 1000))
u = a + b * x
y = np.random.normal(u, d)
```

Preparing samples for distribution as instructed in task.

```python
plt.scatter(x, y)
for i in range(1000):
    x_2 = np.linspace(-6, 6, 100)
    y_2 = b[i] * x_2 + a[i]
    plt.plot(x_2, y_2, 'g', alpha=0.05)
plt.xlim([-6, 6])
plt.title('y ~ N($\mu, \sigma$)')
plt.show()
```
